```python
# Import data for ULX and YSO proposals

ULX = pd.read_csv('ULX.csv')
YSO = pd.read_csv('YSO.csv')

print(type(ULX['Proposal']))
```

```python
# Attemp_1 with SentenceTransformer

model_SentenceTransformer = SentenceTransformer('all-MiniLM-L12-v2')

def create_embeddings_with_sentenceTransformer(proposals):
    embeddings = []
    for proposal in proposals:
        embeddings.append(model_SentenceTransformer.encode(proposal, convert_to_tensor=True));
    return embeddings
```

```python
embeddings_ULX_sentenceTransformer = create_embeddings_with_sentenceTransformer(ULX['Proposal']);
embeddings_YSO_sentenceTransformer = create_embeddings_with_sentenceTransformer(YSO['Proposal']);
```

```python
#Compute cosine-similarities for each keyword collection with every paragraph

def find_keyword_cos_sim(embeddingKeywords, embeddingParagraphs):
  results = []
  for i in range (0,len(embeddingParagraphs)-1):
      results.append(util.cos_sim(embeddingKeywords, embeddingParagraphs[i]))
  return results
```

```python
#Compute cosine-similarities for each sentence with each other sentence

def find_average_cos_sim(embeddings1, embedding2):
  num = 0
  cosine_scores = 0
  for i in range (0,len(embeddings1)-1):
    for j in range (0,len(embedding2)-1):
      cosine_scores += util.cos_sim(embeddings1[i], embedding2[j])
      num+=1
  return cosine_scores/num
```

```python
# Attemp_2 with astrollama
!pip install accelerate

tokenizer = AutoTokenizer.from_pretrained(
    pretrained_model_name_or_path="universeTBD/astrollama"
)

model_astrollama = AutoModelForCausalLM.from_pretrained(
    pretrained_model_name_or_path="universeTBD/astrollama"
)
```

```python
def create_embeddings_with_astrollama(proposals):
    embeddings = []
    proposals.to_dict();
    for proposal in proposals:
        print(proposal)
        inputs = tokenizer(
            proposal,
            return_tensors="pt",
            return_token_type_ids=False,
            padding=True,
            truncation=True,
            max_length=4096
        )
        inputs.to(model_astrollama.device)
        outputs = model_astrollama(**inputs, output_hidden_states=True)

        # Last layer of the hidden states. Get average embedding of all tokens
        embedding = outputs["hidden_states"][-1][:, 1:, ...].mean(1).detach().cpu().numpy()
        embeddings.append(embedding)
    return embeddings
```

```python
ULX_embeddings_astrollama = create_embeddings_with_astrollama(ULX['Proposal'])
YSO_embeddings_astrollama = create_embeddings_with_astrollama(YSO['Proposal'])
```

```python
# ULX/ULX
find_average_cos_sim(ULX_embeddings_astrollama,ULX_embeddings_astrollama)

# ULX/YSO
find_average_cos_sim(ULX_embeddings_astrollama,YSO_embeddings_astrollama)

# YSO/YSO
find_average_cos_sim(YSO_embeddings_astrollama,YSO_embeddings_astrollama)
```
